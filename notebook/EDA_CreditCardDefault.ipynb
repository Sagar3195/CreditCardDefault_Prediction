```python
#Create function that will remove highly correlated features using threshold value
#If the two independent features are highly correlated then we can drop one of the features of the dataframe.
## Removing correlated independent features using threshold value
threshold = 0.5 
```

```python
def correlation(dataset, threshold):
  columns = set() #set of the all name of correlated columns 
  corr_matrix = dataset.corr()
  for i in range(len(corr_matrix.columns)):
    for j in range(i):
      if abs(corr_matrix.iloc[i,j]) > threshold: 
        col_name = corr_matrix.columns[i]  
        columns.add(col_name)
  return columns
```

```python
correlation(df.iloc[:,:-1], threshold)
```
